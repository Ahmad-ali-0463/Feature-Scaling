```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv('D:\summer\ML\covid_toy.csv')
print(df.head())
```

```python
df['cough'].value_counts()
```

```python
df['city'].value_counts()
```

```python
x_test,x_train,y_test,y_train = train_test_split(df.drop(columns= 'has_covid'),
                                                 df['has_covid'],
                                                 random_state=0,
                                              test_size=0.3)
```

```python
si = SimpleImputer()
x_train_fever= si.fit_transform(x_train[['fever']])
x_test_fever= si.fit_transform(x_test[['fever']])

x_train_fever.shape
```

```python
oe = OrdinalEncoder()
x_train_cough=oe.fit_transform(x_train[['cough']])
x_test_cough=oe.transform(x_test[['cough']])
x_train_cough.shape
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(drop='first',sparse_output=False)
x_train_city_gender = ohe.fit_transform(x_train[['gender','city']])
x_test_city_gender = ohe.transform(x_test[['gender','city']])
print(x_train_city_gender.shape)
```

```python
x_train_age = x_train.drop(columns=['gender','fever','cough','city']).values
x_test_age = x_test.drop(columns=['gender','fever','cough','city']).values
x_train_age.shape
```

```python
x_train_transformed = np.concatenate(
    (x_train_age, x_train_cough, x_train_fever, x_train_city_gender),
    axis=1
)

x_test_transformed = np.concatenate(
    (x_test_age, x_test_cough, x_test_fever, x_test_city_gender),
    axis=1
)
x_train_transformed.shape
```

```python
from sklearn.compose import ColumnTransformer
transform = ColumnTransformer(transformers=[('tnf1',SimpleImputer(),['fever']),('tnf2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),('trf3',OneHotEncoder(drop='first',sparse_output=False),['gender','city'])],remainder='passthrough')
transform.fit_transform(x_train).shape
transform.fit_transform(x_test).shape
```
